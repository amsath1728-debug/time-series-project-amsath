```python
# =============================================================================
# ADVANCED CAUSAL INFERENCE: IMPLEMENTING DOWHY FOR TREATMENT EFFECT ESTIMATION
# =============================================================================

# Import necessary libraries
import pandas as pd
import numpy as np
import statsmodels.api as sm
from dowhy import CausalModel  # CORRECTED IMPORT
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

print("=== ADVANCED CAUSAL INFERENCE WITH DOWHY ===")
print("Loading libraries and setting up environment...")

# Set random seed for reproducibility
np.random.seed(42)

# =============================================================================
# TASK 1: DATA SELECTION AND VARIABLE DEFINITION
# =============================================================================

print("\n" + "="*60)
print("TASK 1: DATA SELECTION AND VARIABLE DEFINITION")
print("="*60)

# Load the Lalonde dataset from Statsmodels
print("Loading LaLonde NSW dataset from Statsmodels...")
try:
    lalonde_data = sm.datasets.get_rdataset('lalonde', 'Matching').data
    print("Dataset loaded successfully!")
except Exception as e:
    print(f"Error loading dataset: {e}")
    print("Creating synthetic dataset as fallback...")
    # Create synthetic data as fallback
    np.random.seed(42)
    n = 1000
    lalonde_data = pd.DataFrame({
        'treat': np.random.binomial(1, 0.3, n),
        'age': np.random.normal(35, 10, n),
        'educ': np.random.normal(12, 2, n),
        'black': np.random.binomial(1, 0.3, n),
        'hisp': np.random.binomial(1, 0.1, n),
        'married': np.random.binomial(1, 0.5, n),
        'nodegr': np.random.binomial(1, 0.2, n),
        're74': np.random.exponential(5000, n),
        're75': np.random.exponential(5500, n),
        're78': np.random.exponential(6000, n)
    })

# Display basic dataset information
print(f"Dataset shape: {lalonde_data.shape}")
print("\nFirst few rows of the dataset:")
print(lalonde_data.head())

# Define key variables for causal analysis
print("\n--- VARIABLE DEFINITIONS ---")
print("Treatment: 'treat' (1 = received job training, 0 = control group)")
print("Outcome: 're78' (real earnings in 1978)")
print("Confounders: Pre-treatment variables that may affect both treatment and outcome")

# Selected confounders based on domain knowledge
confounders = ['age', 'educ', 'black', 'hisp', 'married', 'nodegr', 're74', 're75']

print(f"\nSelected confounders: {confounders}")

# Create analysis dataset
analysis_df = lalonde_data[['treat', 're78'] + confounders].copy()

# Data quality checks
print(f"\n--- DATA QUALITY CHECKS ---")
print(f"Missing values:\n{analysis_df.isnull().sum()}")
print(f"\nTreatment group sizes:")
print(f"Control group (treat=0): {len(analysis_df[analysis_df['treat'] == 0])}")
print(f"Treatment group (treat=1): {len(analysis_df[analysis_df['treat'] == 1])}")

# Display summary statistics
print("\n--- SUMMARY STATISTICS ---")
print(analysis_df.describe())

# =============================================================================
# TASK 2: CAUSAL GRAPH FORMULATION AND IDENTIFICATION
# =============================================================================

print("\n" + "="*60)
print("TASK 2: CAUSAL GRAPH AND IDENTIFICATION STRATEGY")
print("="*60)

# Define causal graph using GML syntax
causal_graph = """
digraph {
    // Pre-treatment covariates
    age; educ; black; hisp; married; nodegr; re74; re75;

    // Treatment and outcome
    treat; re78;

    // Causal relationships - confounders affect both treatment and outcome
    age -> treat;
    age -> re78;
    educ -> treat;
    educ -> re78;
    black -> treat;
    black -> re78;
    hisp -> treat;
    hisp -> re78;
    married -> treat;
    married -> re78;
    nodegr -> treat;
    nodegr -> re78;
    re74 -> treat;
    re74 -> re78;
    re75 -> treat;
    re75 -> re78;

    // Causal effect of interest
    treat -> re78;
}
"""

print("Causal Graph Definition Created")

# Create the causal model
print("\nCreating Causal Model with DoWhy...")
try:
    model = CausalModel(
        data=analysis_df,
        treatment='treat',
        outcome='re78',
        graph=causal_graph,
        common_causes=confounders
    )
    print("Causal Model successfully created!")
except Exception as e:
    print(f"Error creating causal model: {e}")
    # Alternative approach without graph
    model = CausalModel(
        data=analysis_df,
        treatment='treat',
        outcome='re78',
        common_causes=confounders
    )
    print("Causal Model created using common causes only")

# Identify the causal effect
print("\n--- IDENTIFICATION PHASE ---")
identified_estimand = model.identify_effect(proceed_when_unidentifiable=True)
print("Identified Estimand:")
print(identified_estimand)

# =============================================================================
# TASK 3: ESTIMATION - MULTIPLE METHODS
# =============================================================================

print("\n" + "="*60)
print("TASK 3: ESTIMATION - MULTIPLE METHODS")
print("="*60)

# Method 1: Propensity Score Weighting (IPW)
print("\n--- METHOD 1: PROPENSITY SCORE WEIGHTING (IPW) ---")
try:
    estimate_ipw = model.estimate_effect(
        identified_estimand,
        method_name="backdoor.propensity_score_weighting",
        target_units="ate"
    )
    print(f"ATE using IPW: ${estimate_ipw.value:,.2f}")
    print(f"95% Confidence Interval: [${estimate_ipw.get_confidence_intervals()[0][0]:,.2f}, ${estimate_ipw.get_confidence_intervals()[0][1]:,.2f}]")
except Exception as e:
    print(f"IPW estimation failed: {e}")
    estimate_ipw = None

# Method 2: Linear Regression
print("\n--- METHOD 2: LINEAR REGRESSION ---")
try:
    estimate_lr = model.estimate_effect(
        identified_estimand,
        method_name="backdoor.linear_regression",
        target_units="ate"
    )
    print(f"ATE using Linear Regression: ${estimate_lr.value:,.2f}")
    print(f"95% Confidence Interval: [${estimate_lr.get_confidence_intervals()[0][0]:,.2f}, ${estimate_lr.get_confidence_intervals()[0][1]:,.2f}]")
except Exception as e:
    print(f"Linear regression estimation failed: {e}")
    estimate_lr = None

# Method 3: Propensity Score Stratification
print("\n--- METHOD 3: PROPENSITY SCORE STRATIFICATION ---")
try:
    estimate_strat = model.estimate_effect(
        identified_estimand,
        method_name="backdoor.propensity_score_stratification",
        target_units="ate"
    )
    print(f"ATE using Propensity Score Stratification: ${estimate_strat.value:,.2f}")
    print(f"95% Confidence Interval: [${estimate_strat.get_confidence_intervals()[0][0]:,.2f}, ${estimate_strat.get_confidence_intervals()[0][1]:,.2f}]")
except Exception as e:
    print(f"Stratification estimation failed: {e}")
    estimate_strat = None

# =============================================================================
# TASK 4: SENSITIVITY ANALYSIS
# =============================================================================

print("\n" + "="*60)
print("TASK 4: SENSITIVITY ANALYSIS")
print("="*60)

print("Performing sensitivity analysis...")

# Simple robustness checks with different confounder sets
print("\n--- ROBUSTNESS CHECKS WITH DIFFERENT SPECIFICATIONS ---")

specifications = [
    ['age', 'educ', 're74', 're75'],
    ['age', 'educ', 'black', 'hisp', 're74', 're75'],
    confounders
]

robustness_results = []
for i, spec in enumerate(specifications, 1):
    try:
        temp_model = CausalModel(
            data=analysis_df,
            treatment='treat',
            outcome='re78',
            common_causes=spec
        )
        temp_estimand = temp_model.identify_effect(proceed_when_unidentifiable=True)
        temp_estimate = temp_model.estimate_effect(temp_estimand, method_name="backdoor.linear_regression")
        robustness_results.append(temp_estimate.value)
        print(f"  Specification {i} ({len(spec)} confounders): ATE = ${temp_estimate.value:,.2f}")
    except Exception as e:
        print(f"  Specification {i} failed: {e}")
        robustness_results.append(None)

# =============================================================================
# TASK 5: VISUALIZATION AND RESULTS
# =============================================================================

print("\n" + "="*60)
print("TASK 5: RESULTS VISUALIZATION")
print("="*60)

try:
    # Create visualizations
    plt.figure(figsize=(15, 10))

    # Plot 1: Distribution of earnings
    plt.subplot(2, 3, 1)
    sns.boxplot(data=analysis_df, x='treat', y='re78')
    plt.title('Earnings by Treatment Group')
    plt.xlabel('Treatment (0=Control, 1=Training)')
    plt.ylabel('1978 Earnings ($)')

    # Plot 2: Prior earnings comparison
    plt.subplot(2, 3, 2)
    sns.boxplot(data=analysis_df, x='treat', y='re74')
    plt.title('1974 Earnings (Pre-treatment)')
    plt.xlabel('Treatment Group')
    plt.ylabel('1974 Earnings ($)')

    # Plot 3: Age distribution
    plt.subplot(2, 3, 3)
    sns.boxplot(data=analysis_df, x='treat', y='age')
    plt.title('Age Distribution')
    plt.xlabel('Treatment Group')
    plt.ylabel('Age')

    # Plot 4: Treatment effect estimates
    plt.subplot(2, 3, 4)
    methods = []
    estimates = []
    cis_lower = []
    cis_upper = []

    if estimate_ipw:
        methods.append('IPW')
        estimates.append(estimate_ipw.value)
        cis_lower.append(estimate_ipw.get_confidence_intervals()[0][0])
        cis_upper.append(estimate_ipw.get_confidence_intervals()[0][1])

    if estimate_lr:
        methods.append('Linear Reg')
        estimates.append(estimate_lr.value)
        cis_lower.append(estimate_lr.get_confidence_intervals()[0][0])
        cis_upper.append(estimate_lr.get_confidence_intervals()[0][1])

    if estimate_strat:
        methods.append('Stratification')
        estimates.append(estimate_strat.value)
        cis_lower.append(estimate_strat.get_confidence_intervals()[0][0])
        cis_upper.append(estimate_strat.get_confidence_intervals()[0][1])

    if methods:
        plt.errorbar(methods, estimates,
                    yerr=[np.array(estimates) - np.array(cis_lower),
                          np.array(cis_upper) - np.array(estimates)],
                    fmt='o', capsize=5, markersize=8)
        plt.axhline(y=0, color='r', linestyle='--', alpha=0.5)
        plt.title('ATE Estimates with Confidence Intervals')
        plt.ylabel('Treatment Effect ($)')

    # Plot 5: Robustness across specifications
    plt.subplot(2, 3, 5)
    if robustness_results:
        spec_labels = [f'Spec {i+1}' for i in range(len(robustness_results)) if robustness_results[i] is not None]
        spec_values = [val for val in robustness_results if val is not None]
        plt.bar(spec_labels, spec_values)
        plt.title('Robustness Across Specifications')
        plt.ylabel('ATE Estimate ($)')

    # Plot 6: Correlation heatmap
    plt.subplot(2, 3, 6)
    correlation_matrix = analysis_df.corr()
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
    plt.title('Variable Correlations')

    plt.tight_layout()
    plt.savefig('causal_inference_results.png', dpi=300, bbox_inches='tight')
    print("Results visualization saved as 'causal_inference_results.png'")

except Exception as e:
    print(f"Visualization failed: {e}")

# =============================================================================
# FINAL RESULTS SUMMARY
# =============================================================================

print("\n" + "="*60)
print("FINAL RESULTS SUMMARY")
print("="*60)

print("ESTIMATED AVERAGE TREATMENT EFFECT (ATE):")
if estimate_ipw:
    print(f"  Primary Estimate (IPW): ${estimate_ipw.value:,.2f}")
    print(f"  95% CI: [${estimate_ipw.get_confidence_intervals()[0][0]:,.2f}, ${estimate_ipw.get_confidence_intervals()[0][1]:,.2f}]")

print(f"\nRobustness Checks:")
if estimate_lr:
    print(f"  Linear Regression: ${estimate_lr.value:,.2f}")
if estimate_strat:
    print(f"  Propensity Score Stratification: ${estimate_strat.value:,.2f}")

# Interpretation
primary_estimate = estimate_ipw if estimate_ipw else estimate_lr
if primary_estimate:
    if primary_estimate.value > 0:
        ci_lower = primary_estimate.get_confidence_intervals()[0][0]
        significance = "statistically significant" if ci_lower > 0 else "suggestive but not statistically significant"
        print(f"\nINTERPRETATION:")
        print(f"The job training program appears to have a {significance} positive effect on earnings.")
        print(f"On average, participants earned ${primary_estimate.value:,.0f} more than non-participants in 1978.")
    else:
        print(f"\nINTERPRETATION:")
        print(f"The analysis does not provide strong evidence for a positive effect of the job training program.")

print(f"\nSENSITIVITY ASSESSMENT:")
print("The analysis includes multiple estimation methods and robustness checks.")
print("Results should be interpreted with caution due to potential unmeasured confounding.")

print("\n" + "="*60)
print("ANALYSIS COMPLETE")
print("="*60)
```
